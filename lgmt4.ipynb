# Importing libraries

```python

import cv2
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
```

# **Reading of the image**

```python
img = cv2.imread("32.jpg")
```

# For showing the image

```python
plt.imshow(img)
b,q,r = cv2.split(img)
img = cv2.merge((r,q,b))
```

# Showing the origninal image

```python
plt.imshow(img)
plt.title(" Original Image")
```

# Conversion of original image into grayscale image

```python
img = mpimg.imread('32.jpg')
```

```python
R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]
imgGray = 0.2989 * R + 0.5870 * G + 0.1140 * B
plt.imshow(imgGray, cmap='gray')
plt.show()
```

# Conversion of grayscale image into inverted image

```python
 img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(8,8))
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

# Soothening of the inverted image

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothened Image")
plt.show()
```

# Final Sketch 

```python
final = cv2.divide(img_gray, 255 - img_smoothing, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch")
plt.show()
```
